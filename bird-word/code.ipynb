```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import regex as re

plt.style.use('fivethirtyeight')
```

```python
with open ("data/lyrics.txt") as text:
    lyrics = text.read().replace('\n',' ').lower()
```

```python
lyrics
```

```python
def post_worder (text):
     
    # remove non-letters
    letters_only = re.sub("[^a-zA-Z]", " ", text)   
    
    # convert post into individual words
    words = letters_only.split()
    
    return(words)
```

```python
words = post_worder(lyrics)
```

```python
cleaned_words = [i for i in words if i != "s"]
```

```python
len(cleaned_words)
```

```python
y = []
x = []
count = 0
birdcount = 0

while count < len(cleaned_words):
    if cleaned_words[count] == "bird":
        birdcount +=1
        count +=1
        y.append(birdcount/count)
        x.append(count)
    else:
        count +=1
        y.append(birdcount/count)
        x.append(count)
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(x=x, y=y)
plt.title("Ratio of bird to word in 'Bird is the Word'")
plt.ylabel("Ratio of bird to word")
plt.xlabel("Number of words\n(Total no of words = 124)")
```
